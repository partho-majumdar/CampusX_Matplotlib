```python
import numpy as np  # type: ignore
import pandas as pd # type: ignore
import matplotlib.pyplot as plt # type: ignore
import seaborn as sns # type: ignore
```

```python
# --------------------------- 2d line plot

''' 
Bivariate analysis -> from multiple column want to analysis with "2 column"
    - Numerical vs Categorical
    - Numerical vs Numerical
    - Time series data
'''
```

```python
price = [48000, 54000, 57000, 49000, 47000, 45000]
year = [2015, 2016, 2017, 2018, 2019, 2020]
plt.plot(year, price) # x axis = first, y axis = second 
```

```python
# from pandas dataframe

batsman = pd.read_csv('./sharma-kohli.csv')
batsman 

plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# multiple plot

plt.plot(batsman['index'], batsman['V Kohli']) 
plt.plot(batsman['index'], batsman['RG Sharma']) 
```

```python
# label and title

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.title('Rohit Sharma vs Virat Kohli run comparison')
plt.xlabel('Season of IPL')
plt.ylabel('Run scored')
```

```python
# plot color 

plt.plot(batsman['index'], batsman['V Kohli'], color = 'green')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black')
plt.title("Rohit Sharma vs Virat Kohli run comparison")
plt.xlabel("Season of IPL")
plt.ylabel("Run scored")
```

```python
# plot color using hex code 

plt.plot(batsman['index'], batsman['V Kohli'], color = '#fa4e09')
plt.plot(batsman['index'], batsman['RG Sharma'], color = '#09d2fa')
plt.title("Rohit Sharma vs Virat Kohli run comparison")
plt.xlabel("Season of IPL")
plt.ylabel("Run scored")
```

```python
# plot line style

plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle = 'dashed')
# plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dotted')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashdot')
plt.title("Rohit Sharma vs Virat Kohli run comparison")
plt.xlabel("Season of IPL")
plt.ylabel("Run scored")
```

```python
# line width

plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle = 'dashed', linewidth = 4)
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashdot', linewidth = 2.5)
plt.title("Rohit Sharma vs Virat Kohli run comparison")
plt.xlabel("Season of IPL")
plt.ylabel("Run scored")
```

```python
# plot marker (i want to see the plot point) and markersize

plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle = 'dashed', linewidth = 1, marker = 'D')
# plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashdot', linewidth = 1, marker = '+')
# plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashdot', linewidth = 1, marker = '.')
# plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashdot', linewidth = 1, marker = '>')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashdot', linewidth = 1, marker = 'o', markersize = 10)
plt.title("Rohit Sharma vs Virat Kohli run comparison")
plt.xlabel("Season of IPL")
plt.ylabel("Run scored")
```

```python
# legend -> guide which line refer which thing (use label and call legend())

plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle = 'dashed', linewidth = 1, marker = 'D', label = 'Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashdot', linewidth = 1, marker = '+', label = 'Rohit')
plt.title("Rohit Sharma vs Virat Kohli run comparison")
plt.xlabel("Season of IPL")
plt.ylabel("Run scored")
plt.legend() # this is default
plt.legend(loc = 'best')
# plt.legend(loc = 'upper right')
# plt.legend(loc = 'upper left')
# plt.legend(loc = 'center')
```

```python
# limiting axis

price = [48000, 54000, 57000, 49000, 47000, 45000, 6500000] # as last value is so high so this is outlier so need to trim 
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
plt.plot(year, price)
plt.ylim(0, 75000)
plt.xlim(2017, 2019)
```

```python
# grid 

plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle = 'dashed', linewidth = 1, marker = 'D', label = 'Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashdot', linewidth = 1, marker = 'o', label = 'Rohit')
plt.title("Rohit Sharma vs Virat Kohli run comparison")
plt.xlabel("Season of IPL")
plt.ylabel("Run scored")
plt.legend()
plt.grid()
```

```python
# show (print graph need to call explicitly for pycharm, flask, website)

plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle = 'dashed', linewidth = 1, marker = 'D', label = 'Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashdot', linewidth = 1, marker = 'o', label = 'Rohit')
plt.title("Rohit Sharma vs Virat Kohli run comparison")
plt.xlabel("Season of IPL")
plt.ylabel("Run scored")
plt.legend()
plt.grid()
plt.show()
```

```python
# --------------------- Bar chart

''' 
Bivariate analysis 
    - Numerical vs Categorical 
    - Aggregate analysis of group
'''
```

```python
# simple bar chart (vertical bar chart)

children = [10, 20, 40, 10, 30]
colors = ['red', 'blue', 'green', 'yellow', 'pink']
plt.bar(colors, children, color = 'green')
plt.title("Children vs Color bar chart")
plt.xlabel("color")
plt.ylabel("number of children")
```

```python
# horizontal bar chart 

plt.barh(colors, children, color = 'black')
plt.title("Children vs Color bar chart")
plt.xlabel("number of children")
plt.ylabel("color")
```

```python
# multiple bar chart

df = pd.read_csv('./batsman_season_record.csv')
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'])
plt.bar(df['batsman'], df['2017'])

# graph overlap so other (in matplotlib need some "jugar")
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width = 0.2, color = 'red')
plt.bar(np.arange(df.shape[0]), df['2016'], width = 0.2, color = 'blue')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width = 0.2, color = 'black')

plt.xticks(np.arange(df.shape[0]), df['batsman']) # replace existing label 
plt.show()
```

```python
children = [10, 20, 40, 10, 30]
colors = ['red red red red red', 'blue blue blue blue blue', 'green green green green green', 'yellow yellow yellow yellow yellow', 'pink pink pink pink pink']

plt.bar(colors, children) # overlap value in x axis
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
# Stacked bar chart

df = pd.read_csv('./batsman_season_record.csv')
plt.bar(df['batsman'], df['2017'], label = '2017')
plt.bar(df['batsman'], df['2016'], bottom = df['2017'], label = '2016')
plt.bar(df['batsman'], df['2015'], bottom = df['2016'] + df['2017'], label = '2015')
plt.legend()
plt.show()
```

```python
# ------------------------ Pie chart
''' 
Univariate / Bivariate analysis
    - Categorical vs Numerical
    - To find contribution on a standard scale
'''
```

```python
# simple pie chart

data = [23, 45, 100, 20, 49]
subject = ['english', 'science', 'math', 'bangla', 'gk']
plt.pie(data, labels = subject)
plt.show()
```

```python
# pie chart from dataset

df = pd.read_csv('./gayle-175.csv')
plt.pie(df['batsman_runs'], labels = df['batsman'])
plt.show()
```

```python
# show percentage

df = pd.read_csv('./gayle-175.csv')
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%')
plt.show()
```

```python
# colors

plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%', colors = ['blue', 'green', 'yellow', 'pink', 'cyan', 'brown'])
plt.show()
```

```python
# explode shadow

plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%', explode = [0.1, 0, 0, 0, 0, 0.1], shadow = True)
plt.show()
```
