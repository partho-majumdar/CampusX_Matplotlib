```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# colored scatter plot

iris = pd.read_csv("./iris.csv")
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])
plt.xlabel("Sepal length")
plt.ylabel("Petal length")
```

```python
iris['Species'] = iris['Species'].replace({
    'Iris-setosa': 0,
    'Iris-versicolor': 1,
    'Iris-virginica': 2
})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'])
plt.xlabel("Sepal length")
plt.ylabel("Petal length")
```

```python
# for changing color 
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'viridis')
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'summer')
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'jet')
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'magma')
plt.xlabel("Sepal length")
plt.ylabel("Petal length")
```

```python
# colorbar -> which color represent which value 

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'jet')
plt.xlabel("Sepal length")
plt.ylabel("Petal length")
plt.colorbar()
```

```python
# alpha -> control graph opacity

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'jet', alpha = 0.3)
plt.xlabel("Sepal length")
plt.ylabel("Petal length")
plt.colorbar()
```

```python
# plot size 

plt.figure(figsize = (15, 5)) # width = 15, height = 5 (applicable for all graph)
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'jet', alpha = 0.3)
plt.xlabel("Sepal length")
plt.ylabel("Petal length")
plt.colorbar()
```

```python
# ------------------- Annotations (place a text in a graph)

batters = pd.read_csv('./batter.csv')
sample_df = batters.head(100).sample(25, random_state = 5)
# sample_df = batters.head(100)
plt.scatter(sample_df['avg'], sample_df['strike_rate']) # here don't know which point represent which batsman use annotation
```

```python
x = [1, 2, 3, 4]
y = [5, 6, 7, 8]

plt.scatter(x, y)
plt.text(1, 5, 'point1')
plt.text(2, 6, 'point2')
plt.text(3, 7, 'point3')
plt.text(4, 8, 'point4', fontdict = {
    'size': 12,
    'color': 'brown'
})
```

```python
plt.figure(figsize = (18, 9))
plt.scatter(sample_df['avg'], sample_df['strike_rate'])
plt.xlabel('avg')
plt.ylabel('strike_rate')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
# according to run point size decide 

plt.figure(figsize = (18, 9))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s = sample_df['runs'])
plt.xlabel('avg')
plt.ylabel('strike_rate')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
# Horizontal (axhline) vs Vertical (axvline) lines

plt.figure(figsize = (18, 9))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s = sample_df['runs'])
plt.xlabel('avg')
plt.ylabel('strike_rate')


for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])

plt.axhline(130, color = 'red')
plt.axhline(140, color = 'green')
plt.axvline(30, color = 'red')
```

```python
# --------------------- Subplots

# second way of plotting graph


# plt.subplots() # give me two default value --> fig(graph) & axis
fig, ax = plt.subplots(figsize = (15, 9))
ax.scatter(batters['avg'], batters['strike_rate'], color = 'red', marker = '+')
ax.set_title("Avg vs strike rate graph")
ax.set_xlabel("Avg")
ax.set_ylabel("Strike-rate")
fig.show()
```

```python
# draw side by side two multiple graph (avg vs strike rate) & (runs vs strike rate)

batters.head()
fig, ax = plt.subplots(nrows = 2, ncols = 1, figsize = (20, 10), sharex = True) # sharex = True (only show value in x co-ordinate)
ax[0].scatter(batters['avg'], batters['strike_rate'], color = 'red', marker = '+')
ax[1].scatter(batters['avg'], batters['runs'], color = 'green')

ax[0].set_title("Avg vs strike rate")
ax[0].set_xlabel("Avg")
ax[0].set_ylabel("Strike rate")

ax[1].set_title("Avg vs runs")
ax[1].set_xlabel("Avg")
ax[1].set_ylabel("total runs")

fig.show()
```

```python
# 4 box with 2 row, and 2 col
''' 
2D array 
    - (0, 0) -> first box 
    - (0, 1) -> second box 
    - (1, 0) -> third box 
    - (1, 1) -> fourth box  
'''

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (20, 10))

ax[0, 0].scatter(batters['avg'], batters['strike_rate'], color = 'red')
ax[0, 1].scatter(batters['avg'], batters['runs'], color = 'green')
ax[1, 0].hist(batters['avg'], color = 'pink')
ax[1, 1].hist(batters['runs'], color = 'brown')
```

```python
# Second way to do the same thing

fig = plt.figure(figsize = (20, 10))

ax1 = fig.add_subplot(2, 2, 1) # 2 row, 2 col, 1st graph
ax1.scatter(batters['avg'], batters['strike_rate'], color = 'red')

ax2 = fig.add_subplot(2, 2, 2) # 2 row, 2 col, 2nd graph
ax2.hist(batters['runs'], color = 'brown')

ax3 = fig.add_subplot(2, 2, 3) # 2 row, 2 col, 3rd graph
ax3.scatter(batters['avg'], batters['runs'], color = 'green')

ax4 = fig.add_subplot(2, 2, 4) # 2 row, 2 col, 4th graph
ax4.hist(batters['avg'], color = 'pink')
```

```python
# -------------------- 3D graph

fig = plt.figure(figsize = (20, 10))
ax = plt.subplot(projection = '3d')
ax.scatter3D(batters['runs'], batters['avg'], batters['strike_rate'])

ax.set_title("IPL batsman run analysis")
ax.set_xlabel("runs")
ax.set_ylabel("avg")
ax.set_zlabel("strike_rate")
```

```python
# 3D line plot

x = [0, 1, 5, 25]
y = [0, 10, 13, 0]
z = [0, 13, 20, 9]

fig = plt.figure()
ax = plt.subplot(projection = '3d')

ax.scatter3D(x, y, z, s = [100, 100, 100, 100])
ax.plot3D(x, y, z)
```

```python
# ------------------------ 3D Surface plot

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x, y)
z = (xx ** 2) + (yy ** 2)

fig = plt.figure(figsize = (20, 10))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx, yy, z, cmap = 'viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize = (20, 10))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx, yy, z, cmap = 'viridis')
fig.colorbar(p)
```

```python
# Contour graph -> represent 3d graph into a 2d graph

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x, y)
z = (xx ** 2) + (yy ** 2)

fig = plt.figure(figsize = (10, 8))
ax = plt.subplot()
p = ax.contour(xx, yy, z, cmap = 'viridis')
fig.colorbar(p)
```

```python
# contourf -> filled contour

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x, y)
z = (xx ** 2) + (yy ** 2)

fig = plt.figure(figsize = (10, 8))
ax = plt.subplot()
p = ax.contourf(xx, yy, z, cmap = 'viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize = (10, 8))
ax = plt.subplot()
p = ax.contourf(xx, yy, z, cmap = 'viridis')
fig.colorbar(p)
```

```python
# Heatmap -> how value is distributed (compare 2 categorical data)

delivery = pd.read_csv('./IPL_Ball_by_Ball_2008_2022.csv')
temp_df = delivery[(delivery['ballnumber'].isin([1, 2, 3, 4, 5, 6])) & (delivery['batsman_run'] == 6)]

grid = temp_df.pivot_table(index = 'overs', columns = 'ballnumber', values = 'batsman_run', aggfunc = 'count')
plt.figure(figsize = (10, 8))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(), list(range(1, 21)))
plt.xticks(np.arange(0, 6), list(range(1, 7)))
plt.colorbar()
```
